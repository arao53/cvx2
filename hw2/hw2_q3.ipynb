```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the function

def f(x):
    if x <= 0:
        return -2*x
    elif x < 4:
        return -0.5*x
    else:
        return x - 6
xd = np.linspace(-2, 6, 100)
yd = [f(x) for x in xd]
```

```python
fig, ax = plt.subplots()
ax.plot(xd, yd)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Piecewise function')
ax.set(xlim=(-2, 6), ylim=(-4, 4), aspect=1)
plt.show()
fig.savefig('hw2_q3a.png')
```
