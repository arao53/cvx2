```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
iters = 100

A = np.array([[0, 1, -1], [-1, 0, 1], [1, -1, 0]])
A1 = A[:,0]
A2 = A[:,1]
A3 = A[:,2]

s = np.array([1, -1, 0])

x = np.ndarray((iters, 3))
nu = np.ndarray((iters, 3))
phi_i = np.ndarray((iters, 3))
phi_m = np.ndarray((iters, 1))
alpha = 0.1
```

```python
nu[0, :] = np.zeros(3)

for idx in range(iters-1):
    x[idx, 0] = -0.5 * nu[idx, :].T @ A1
    x[idx, 1] = -0.5 * nu[idx, :].T @ A2
    x[idx, 2] = -0.5 * nu[idx, :].T @ A3

    nu[idx+1, :] = nu[idx, :] - alpha * (A.T @ A @ nu[idx, :] - s)
    
```

```python
fig, ax = plt.subplots()
ax.plot(x[:-1,:], alpha = 0.5)
ax.legend(labels = ['x1', 'x2', 'x3'])
ax.set(xlabel='iteration', ylabel='potentials [V]')
```
